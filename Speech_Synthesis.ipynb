```python
from transformers import AutoProcessor, SeamlessM4Tv2Model
import torchaudio

processor = AutoProcessor.from_pretrained("facebook/seamless-m4t-v2-large")
model = SeamlessM4Tv2Model.from_pretrained("facebook/seamless-m4t-v2-large")

# from text
text = '''subah uth kar nashta tayar kiya or phir kaya
university k liye gaya lekin udher class cancel thi
university mai dosto k sath chess khaila
sham ko ghar wapis aya tu mehman aye howe thy tu sabse mila
mehmano k liye khana pakaya
raat ko mehmano k sath bohat gup shup lagayi or phir sogaya'''
text_inputs = processor(text = text, src_lang="urd", return_tensors="pt")
# text_inputs = processor(text = "نادرا نے چائے والا کے نام سے مشہور ارش د خان کا شناختی کارڈ کیوں بلاک کیا", src_lang="urd", return_tensors="pt")
audio_array_from_text = model.generate(**text_inputs, tgt_lang="urd")[0].cpu().numpy().squeeze()


from IPython.display import Audio

sample_rate = model.config.sampling_rate
Audio(audio_array_from_text, rate=sample_rate)
```
